```python
import os
import pandas as pd 
def read_data():
    file_directory = ''
    file_name = 'pima-indians-diabetes.data.csv'
    file_path = os.path.join(file_directory,file_name)
    try:
        df = pd.read_csv(file_path,header=None,names=[])
        return(df)
    except FileNotFoundError as e:
        print('FileNotFoundError',e)

df = read_data()
```
